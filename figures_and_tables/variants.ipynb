```python
import pandas as pd
import ast
```

```python
file_path = '../predictions/unsolved_retnet.tsv'

data = []

with open(file_path, 'r') as file:
    for line in file:
        # Use ast.literal_eval to safely evaluate the string as a list
        entry = ast.literal_eval(line.strip())
        data.append(entry)

# Process the data
processed_data = []

for entry in data:
    position = entry[0]
    retina = entry[1]
    gtex = entry[2]
    gtex2 = entry[3]

    number_of_predictions = len(retina)
    for i in range(number_of_predictions):
        out = [position]
        for j in retina[i].split('|'):
            out.append(j)
        for k in gtex[i].split('|')[2:]:
            out.append(k)
        for l in gtex[i].split('|')[2:]:
            out.append(l)
        processed_data.append(out)


predictions = pd.DataFrame(processed_data, columns=['Position', 'Base', 'Transcript', 
                                                    'DS_AG_retina', 'DS_AL_retina', 'DS_DG_retina', 'DS_DL_retina',
                                                    'DP_AG_retina', 'DP_AL_retina', 'DP_DG_retina', 'DP_DL_retina', 
                                                    'DS_AG_gtex', 'DS_AL_gtex', 'DS_DG_gtex', 'DS_DL_gtex',
                                                    'DP_AG_gtex', 'DP_AL_gtex', 'DP_DG_gtex', 'DP_DL_gtex',
                                                    'DS_AG_gtex2', 'DS_AL_gtex2', 'DS_DG_gtex2', 'DS_DL_gtex2',
                                                    'DP_AG_gtex2', 'DP_AL_gtex2', 'DP_DG_gtex2', 'DP_DL_gtex2',])
print(len(predictions))
predictions.head()
```

```python
for column in ['DS_AG', 'DS_AL', 'DS_DG', 'DS_DL']:
    predictions[column + '_retina'] = pd.to_numeric(predictions[column + '_retina'], errors='coerce')
    predictions[column + '_gtex'] = pd.to_numeric(predictions[column + '_gtex'], errors='coerce')
    predictions[column + '_gtex2'] = pd.to_numeric(predictions[column + '_gtex2'], errors='coerce')

threshold = 0.15

# Filter the DataFrame
only_retina = predictions[(predictions['DS_AG_retina'] > threshold) & (predictions['DS_AG_gtex'] < threshold) |
                          (predictions['DS_AL_retina'] > threshold) & (predictions['DS_AL_gtex'] < threshold) |
                          (predictions['DS_DG_retina'] > threshold) & (predictions['DS_DG_gtex'] < threshold) |
                          (predictions['DS_DL_retina'] > threshold) & (predictions['DS_DL_gtex'] < threshold)]

# Filter the DataFrame
only_retina2 = predictions[(predictions['DS_AG_retina'] > threshold) & (predictions['DS_AG_gtex2'] < threshold) |
                          (predictions['DS_AL_retina'] > threshold) & (predictions['DS_AL_gtex2'] < threshold) |
                          (predictions['DS_DG_retina'] > threshold) & (predictions['DS_DG_gtex2'] < threshold) |
                          (predictions['DS_DL_retina'] > threshold) & (predictions['DS_DL_gtex2'] < threshold)]

print(len(only_retina.drop_duplicates()))
print(len(only_retina2.drop_duplicates()))
```
